```python
# 리스트 초기화
x_train = []
y_train = []

# x가 1~10까지를 생성
for x in range(1,11):
    x_train.append(x)
    y_train.append(2*x+1)

print(x_train)
print(y_train)
```

### keras로 신경망(NN)을 구성

딥러닝의 앞의 레이어의 출력을 다음 레이어의 입력으로 자동으로 연결하고 싶을 때는 Sequentail를 사용 <br>
예를들어, input 3개, 다음 레이어 2개가 있는 경우, input 레이어와 다음 레이어 2개의 하나도 빠짐없이 모든 노드와 연결할 때 Danse 사용 <br>

```python
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import Adam
import numpy as np # Import numpy

# Convert lists to NumPy arrays
x_train_np = np.array(x_train)
y_train_np = np.array(y_train)

# 1. 모델 생성
model = Sequential()              # Sequential모델로 모델을 구성
model.add(Input(shape=(1,)))      # 입력 레이어의 shape을 정의
model.add(Dense(1))               # 모델안에 하나의 레이어를 추가

# 2. 옵티마이저 설정
adam = Adam(learning_rate=0.01, beta_1=0.99, beta_2=0.99)

# 3. 옵티마이저와 손실함수를 모델에 설정
model.compile(loss='mse', optimizer=adam)

# 4. 학습(fit)
hist_model = model.fit(x_train_np, y_train_np, epochs=100, verbose=2)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.optimizers import SGD
import numpy as np # Import numpy

# Convert lists to NumPy arrays
x_train_np = np.array(x_train)
y_train_np = np.array(y_train)

# 1. 모델 생성
model2 = Sequential()              # Sequential모델로 모델을 구성
model2.add(Input(shape=(1,)))      # 입력 레이어의 shape을 정의
model2.add(Dense(1))               # 모델안에 하나의 레이어를 추가

# 2. 옵티마이저 설정
sgd = SGD(learning_rate=0.001)

# 3. 옵티마이저와 손실함수를 모델에 설정
model2.compile(loss='mse', optimizer=sgd)

# 4. 학습(fit)
hist_model2 = model2.fit(x_train_np, y_train_np, epochs=100, batch_size=4, verbose=2)
```
